```python
# Definition for singly-linked list.
from typing import Optional


class ListNode:
    def __init__(self, x):
        self.val = x
        self.next = None

class Solution:
    def hasCycle(self, head: Optional[ListNode]) -> bool:
        list_node = []
        loop : bool = True
        temp_node = head
        if head:
          while(loop):
            if temp_node not in list_node:
              list_node.append(temp_node)
            else:
              return True
            if temp_node.next == None:
              return False
            temp_node = temp_node.next
          else:
            return False
        
```
