```python
# Definition for a binary tree node.
from typing import List, Optional


class TreeNode:
  def __init__(self, val=0, left=None, right=None):
      self.val = val
      self.left = left
      self.right = right
class Solution:
  
  def medium_order(self, root: Optional[TreeNode], elems_list : List):
    if root == None:
      return 
    self.medium_order(root.left)
    elems_list.append(root.val) 
    self.medium_order(root.right)
    
  def getMinimumDifference(self, root: Optional[TreeNode]) -> int:
    elems_list = []
    self.medium_order(root, elems_list)
    result = abs(elems_list[0] - elems_list[1])
    for x in range(0, len(elems_list) - 1):
      temp_min = abs(elems_list[x] - elems_list[1 + x])
      if result >= temp_min:
        result = temp_min
    
    return result
    
```
