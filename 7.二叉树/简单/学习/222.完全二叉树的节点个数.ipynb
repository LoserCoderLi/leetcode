```python
from typing import Optional
# Definition for a binary tree node.
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right
        
class Solution:
    def countNodes(self, root: Optional[TreeNode]) -> int:
      if root == None:
          return 0
      
      left = root.left
      right = root.right
      left_number = 1
      right_number = 1
      while(left):
          left = left.left
          left_number += 1
      while(right):
          right = right.right
          right_number += 1
      
      if left_number == right_number:
          return 2**left_number - 1
      
      left_num = self.countNodes(root.left)
      right_num = self.countNodes(root.right)
      
      node_num = left_num + right_num + 1
      
      return node_num
            
      
      
```
